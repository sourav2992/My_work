# Variance Approach 

## Numerical Variables 

```python
data.describe()
```

```python
numeric_data = data.select_dtypes(include = [np.number])
numeric_data.shape
```

```python
categorical_data = data.select_dtypes(exclude = [np.number])
categorical_data.shape
```

```python
variance = numeric_data.var()
variance
```

```python
numeric_columns = numeric_data.columns
numeric_columns
```

```python
Low_Variance_Col = []

for i in range(0, len(numeric_columns)):
    if variance[i] ==0:
        Low_Variance_Col.append(numeric_data.columns[i])

Low_Variance_Col        
```

```python
numeric_data.shape
```

```python
numeric_data.drop(columns = Low_Variance_Col, axis = 1, inplace = True)
```

# Correlation Approach 

```python
cor_matrx = numeric_data.corr().abs()
cor_matrx
```

```python
upr_tri = cor_matrx.where(np.triu(np.ones(cor_matrx.shape), k=1).astype(np.bool))
upr_tri
```

```python
upr_tri.columns
```

```python
High_corr_Cols_to_drop = [i for i in upr_tri.columns if any(upr_tri[i]>.95)]
High_corr_Cols_to_drop
```

```python
numeric_data.drop(columns = High_corr_Cols_to_drop, axis = 1, inplace = True)
```

# PCA 

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
numeric_data.isnull().sum().sum()/np.product(numeric_data.shape)
```

```python
numeric_data = numeric_data.fillna(numeric_data.mean())
```

```python
numeric_data
```

```python
x=numeric_data.values
x
```

```python
x=scale(x)
x
```

```python
pca = PCA()
pca.fit(x)
```

```python
list(pca.explained_variance_ratio_*100)
```

```python
list((pca.explained_variance_ratio_*100).cumsum())
```

```python
pca1 = PCA(n_components = 5)
pca1.fit(x)
list((pca1.explained_variance_ratio_*100).cumsum())
```
